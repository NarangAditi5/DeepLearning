```python
import tensorflow as tf
print(tf.__version__)
```

```python
a=tf.constant(5)
print(a)

v=tf.constant([1,2,3,4])
print(v)

m = tf.constant([[1,2], [3,4]])
print(m)

t3=tf.constant([[[1,2],
                [3,4]],
                [[5,6],
                [7,8]]])
print(t3)
```

```python
a = tf.constant([10, 20, 30])
b = tf.constant([50, 60, 70])
print(a+b)
```

```python
a = tf.constant([10, 20, 20])
print(a.dtype, a.shape, tf.rank(a))
```

```python
import numpy as np
np_array = np.array([[1,2,3], [4,5,6]])

tensor = tf.convert_to_tensor(np_array)
print(tensor)
```
